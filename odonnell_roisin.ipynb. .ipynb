### Let's analyze tendencies in the Titanic Data graphically:

Show graphically if there is a tendency in the # of siblings / spouses aboard the Titanic over the age of the passengers. 

```python
sorting= df.groupby('Age').SibSp.nunique()
print(sorting.head(3))
plt.plot(sorting, linestyle = ':', color = 'slateblue', marker = 'o',markerfacecolor = 'black')
plt.title('tendancy between # of siblings and spouses and age')
plt.ylabel('# of siblings and spouses')
plt.xlabel('Passenger Ages')
plt.show()
#conclusions
#the younge the passenger, the more likely they are to have more than 2 SibSp
```

### Now you have freedom to create your own plot:

Think in some data analysis that we have not done yet in this jupyter notebook. Plot the most adequate graph to represent such analysis and extract conclusions about it. 


```python
fig1 = plt.figure()
chrt = 0
classes = 0

for i, group in df.groupby('Pclass'):
    chrt += 1 
    classes += 1
    ax = fig1.add_subplot(2,2, chrt)
    group.boxplot('Age',ax=ax, patch_artist=True)
    ax.set_title(classes)
    plt.subplots_adjust(top = 1, wspace = 1, hspace = 0.5)
    ax.set_ylabel('Ages')
    ax.set_ylim(-1, 90)
    
#conclusions
#the median for first class is mid 30s and the class with the most variance is third (it's median is 20s but it's outliers stretch late 70s).
```
